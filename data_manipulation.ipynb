Put stressors and colony sheets together because that data have common columns

```python
#import needed libraries
import pandas as pd

#create dfs
stressors = pd.read_excel('stressors_clean.xlsx')
stressors
```

```python
colony = pd.read_excel('colony_clean.xlsx')
colony
```

```python
# colony has 46 more rows
# do an outer merge and then search for NaN to find out what rows are missing in stressors
# merge on 'State' and 'Year'
outer_merge = pd.merge(colony, stressors, on = (['State', 'Year']), how = 'outer')
outer_merge.tail(50)
```

```python
# Q2 of 2021 doesn't have any stressors data so I will do an inner join and keep that as my merged data
inner_merge = pd.merge(colony, stressors, on = (['State', 'Year']), how = 'inner')
inner_merge
```

```python
# remove 'Other' row
inner_merge = inner_merge[inner_merge.State != 'Other']
inner_merge
```

```python
inner_merge.reset_index(drop = True)
```

```python
#create a new dataframe that replaces 'State' with 'Region' then merge on all columns but state and region.

region = inner_merge['State'].replace(['Illinois', 'Indiana', 'Iowa', 'Ohio', 'Missouri', 'Michigan', 'Minnesota', 'Wisconsin', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas'], 'Midwest')
```
